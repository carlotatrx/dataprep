```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams.update({
    "figure.figsize": (8, 5),
    "axes.grid": True,
    "axes.axisbelow": True
})
```

```python
df = pd.read_csv("sef_series_summary_v2.csv")
df.head()
```

```python
df["variable"].value_counts()
```

```python
bins = np.arange(0, df["span_days"].max() + 365, 365)  # yearly bins

bins_log = np.logspace(
    np.log10(df["span_days"].min()),
    np.log10(df["span_days"].max()),
    20
)
```

```python
variables = sorted(df["variable"].unique())

for var in variables:
    sub = df[df["variable"] == var]

    plt.figure()
    plt.hist(sub["span_days"], bins=bins_log)
    
    plt.xlabel("Series length [days]")
    plt.ylabel("Number of series")
    plt.title(f"Distribution of record lengths  {var}")
    
    # Uncomment if using log bins
    plt.xscale("log")

    plt.tight_layout()
    plt.show()
```
